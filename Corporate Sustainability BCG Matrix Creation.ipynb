```python
import altair as alt #this is what the previous line allows me to access
#this library altair does really nice data visualizations
#I gave it the nickname alt for easy of typing
```

```python
#BCG Matrix using Yahoo Size Scale
alt \
  .Chart(df) \
  .mark_circle() \
  .encode(#allows me to specify details of the graph
    x = alt.X('Market Share:Q', title = '2019 Market Share (%)'), #data used for x variable
    y = alt.Y('Growth Rates:Q', title = 'Past 5 Years Earnings Growth Rate'), #data used for y
    color = 'Company Name', #color codes each data point by company name
    tooltip = ['Company Name', 'Yahoo Rating', 'Revenues','Market Share', 'Growth Rates' ],
    size = 'Yahoo Rating'#determines the circle's size
) .properties( #makes the chart easier to see in these sizes
    width = 350,
    height = 350
)
```

```python
# Revenues v. Yahoo ESG Scores w/Market Share as the size
alt \
  .Chart(df) \
  .mark_circle() \
  .encode(
    #specify x and y now:
    x = alt.X('Revenues:Q', axis = alt.Axis(format='$'), title = '2019 Revenues in Million USD'),
    y = alt.Y('Yahoo Rating:Q', scale = alt.Scale(zero = False)),
    color = 'Company Name', #color codes each data point by company ticker
    tooltip = ['Company Name', 'Yahoo Rating', 'Revenues','Market Share', 'Growth Rates' ],
    size = 'Market Share'#the respective product markets of each company (varies)
) .properties(
    width = 350,
    height = 350
)
```

```python
#BCG Matrix using MSCI Size Scale
alt \
  .Chart(df) \
  .mark_circle() \
  .encode(
    #specify x and y now:
    x = alt.X('Market Share:Q', title = '2019 Market Share (%)'),
    y = alt.Y('Growth Rates:Q', title = 'Past 5 Years Earnings Growth Rate'),
    color = 'Company Name', #color codes each data point by company name
    tooltip = ['Company Name', 'MSCI Rating', 'Revenues','Market Share', 'Growth Rates' ],
    size = 'MSCI Numbers'
) .properties(
    width = 350,
    height = 350
)
#7 = AAA
#1 = CCC
```
